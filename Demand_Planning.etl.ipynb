```python
from IPython import get_ipython
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
```

```python
# Read Files comming from https://www.kaggle.com/utathya/future-volume-prediction
# create a panda Data Frame for each of them
input_dir = './data/'
demo        = pd.read_csv(input_dir + 'demographics.csv')
event       = pd.read_csv(input_dir + 'event_calendar.csv')
historical  = pd.read_csv(input_dir + 'historical_volume.csv')
soda        = pd.read_csv(input_dir + 'industry_soda_sales.csv')
industry    = pd.read_csv(input_dir + 'industry_volume.csv')
price       = pd.read_csv(input_dir + 'price_sales_promotion.csv')
weather     = pd.read_csv(input_dir + 'weather.csv')
forecast    = pd.read_csv(input_dir + 'volume_forecast.csv')
```

```python
# Change the date format of all field format
# It is key in time series
event['YearMonth']  = pd.to_datetime(event['YearMonth'],format='%Y%m')
historical['YearMonth'] = pd.to_datetime(historical['YearMonth'],format='%Y%m')
soda['YearMonth'] = pd.to_datetime(soda['YearMonth'],format='%Y%m')
industry['YearMonth'] = pd.to_datetime(industry['YearMonth'],format='%Y%m')
price['YearMonth'] = pd.to_datetime(price['YearMonth'],format='%Y%m')
weather['YearMonth'] = pd.to_datetime(weather['YearMonth'],format='%Y%m')
```

```python
# Merge the Files to get one dataset including the fields of each csv provided as input.
df = historical.merge(price,on=['Agency','SKU','YearMonth'],how='left')
df = df.merge(soda,on=['YearMonth'],how='left')
df = df.merge(industry,on='YearMonth',how='left')
df = df.merge(event,on=['YearMonth'],how='left')
# Weather is related to Agency only so we mege it to agency 
agency = weather.merge(demo,on=['Agency'],how='left')
df = df.merge(agency,on=['YearMonth','Agency'],how='left')
df.sort_values(by=['YearMonth','Agency','SKU'])
```

```python
# Saving the File to use for the next step
df.to_csv('dataset.csv', index=False)
```
